# Intro to Pandas

Pandas is a library that helps organize and understand information like this into dataframes. These dataframes are kind of like lists or 2d arrays that you see in Python, but they are much more advanced.

students.png

```python
# Run the cell to see how a dataframe gets created.
import pandas as pd

my_dataframe = pd.DataFrame(
    {
        "Name": ("Rei", "Alexis", "Riley", "Anna"),
        "Favorite Color": ("Green", "Yellow", "Purple", "Orange"),
        "Age": (13, 15, 8, 10)
    }
)

print(my_dataframe)
```

```python
# Edit the dataframe to add a column for Favorite Candy.
my_dataframe = pd.DataFrame(
    {
        "Name": ("Rei", "Alexis", "Riley", "Anna"),
        "Favorite Color": ("Green", "Yellow", "Purple", "Orange"),
        "Age": (13, 15, 8, 10),
        # Add favorite candy here:
        "Favorite Candy": ("Snickers", "Twix", "Twizzlers", "None")
    }
)

print(my_dataframe)
```

# Viewing Data
As you can see you can print out an entire dataframe using the print function. However, imagine that you have a dataframe with thousands of rows in it. Printing out that whole dataframe would be an issue.

Pandas has two functions to help with this: head and tail.
* `head()` will show you the first five rows of data.

* `tail()` will show you the last five rows of data.

```python
test_data = pd._testing.makeDataFrame()
# Call the head() function here
test_data.head()
```

```python
# Call the tail() function here
test_data.tail()
```

# Manipulating dataframes

Now that you know how to view data, you can drop or add columns from a dataframe.

```python
test_data = test_data.drop(['A'], axis=1)

test_data.head()
```

```python
# Drop column C here

# Use the head() function to see the first five rows with the new changes.
test_data = test_data.drop(['C'], axis=1)

test_data.head()
```

# Importing Data

Another useful things to do with pandas is to import data into a dataframe so that you can manipulate it. You can read data from any online open source location.

```python
penguins = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')
# Use the head() function to see the first 5 rows of the penguin data.
penguins.head()
```

# Dropping Useless Rows

When you import data like this, sometimes you get bad rows where there isn't useful data. You can use a function called `dropna()` to get rid of bad data.

```python
# Use dot notation to call dropna() on the penguins dataframe to drop the rows where at least one element is mising.
penguins = penguins.dropna()
# Use the head() function to see the first 5 rows of the data.
penguins.head()
```

# Selecting Data

You might also be in a situation where you want to create a dataset with only some of the data in it. In this case, you can select different parts of the data.

```python
# Add code in the cell below to see which island each penguin is from.
print(penguins['island'])
```

```python
# Add code in the cell below to see all penguins species.
print(penguins['species'])
```

You can also select a series of rows using the ` : `.

```python
# Add the code to select rows 14 through 23
print(penguins[14:23])
```

```python
# Add the code to select rows 4 through 15
print(penguins[4:15])
```

# Sorting

You can also sort data by it's value. This can be useful if you want to find specific rows of data, or if you want the data in a certain order for some reason.

```python
# Sort penguines by flipper length
penguins = penguins.sort_values(by="flipper_length_mm", ascending=False)

# Add code to print the first 5 rows
penguins.head()
```

```python
# Sort penguins by body mass
penguins = penguins.sort_values(by="body_mass_g", ascending=False)
# Add code to print the last 5 rows
penguins.tail()
```
