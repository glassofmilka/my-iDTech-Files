# Character Class
Your project is a battle arena where characters fight with class. The winner gets the glory, and the loser will sleep with the fish. To begin, you'll create a class for your character and create the player class methods.

player.png

```python
# Create a character class for your arena players.
class character():  
    # Constructor
    def __init__(self, name, strength, health, defense):
        self.name = name
        self.strength = strength
        self.health = health
        self.defense = defense
    # take_damage() Method
    def take_damage(self,damage):
        damage_taken = damage - self.defense
        self.health -= damage_taken
        return damage_taken
    # attack() Method
    def attack(self,target):
        damage = self.strength * 2
        damage_dealt = target.take_damage(damage)
        return damage_dealt

    # is_alive() Method
    def is_alive(self):
        return self.health > 0
    # print_stats Method
    def print_stats(self):
        print("Name: " + str(self.name))
        print("Strength: " + str(self.strength))
        print("Health: " + str(self.health))
        print("Defense: " + str(self.defense))
```

# Rogue Class
Rogues are fast, and they rely on their dexterity in combat to get the upper hand on their opponent. You're going to create a Rogue sub-class which will use the attack method to give the opponent a chance to gain a critical hit.

In addition, you'll override the **attack()** method to give the Rogue character unique abilities, such as randomly generating a critical hit that causes more damage.

```python
# Import the random library
import random
```

```python
# Create a rogue character class by extending the character class.
class Rogue(character):
    def __init__(self, name, strength, health, defense, dexterity):
        self.name = name
        self.strength = strength
        self.health = health
        self.defense = defense
        self.dexterity = dexterity
    # attack() Method
    def attack(self,target):
        critical_hit = random.randint(0, 50) < self.dexterity
        damage = self.strength * 2
        if critical_hit:
            damage *= 2
        print("***CRITICAL HIT***")
        damage_dealt = target.take_damage(damage)
        return damage_dealt
    def print_stats(self):
        print("Name: " + self.name)
        print("Strength: " + str(self.strength))
        print("Health: " + str(self.health))
        print("Defense: " + str(self.defense))
        print("Dexterity: " + str(self.dexterity))
```

# Arena Battle
Now that you have characters that can attack and take damage, it's time to make them fight in the Arena!

```python
# Create an arena battle.
print(player1.name + " vs. " + player2.name)
while player1.is_alive() and player2.is_alive():
    print(player1.name + ": " + str(player1.health))
    print(player2.name + ": " + str(player2.health))
    damage = player1.attack(player2)
    print(player1.name + " hits " + player2.name + " for " + str(damage))
    damage = player2.attack(player1)
    print(player2.name + " hits " + player1.name + " for " + str(damage))
```
